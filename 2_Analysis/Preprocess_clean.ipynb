# Step 3: Remove strings that come as a result of bad conversion/translation error

```python
def remove_non_ascii(text):
    """
        text: a string
        
        return: modified initial string with only ASCII characters
    """
    if pd.isnull(text):
        return ' '    #to avoid 'float object is not iterable' error
    return ''.join(i for i in text if ord(i)<128)
```

```python
df.assignee = df.assignee.apply(remove_non_ascii)
df.abstract = df.abstract.apply(remove_non_ascii)
df.title = df.title.apply(remove_non_ascii)
df.claims = df.claims.apply(remove_non_ascii)
df.inventor = df.inventor.apply(remove_non_ascii)
df.examiners = df.examiners.apply(remove_non_ascii)
```

# Step 4: create new columns for pieces of CPCs and full citations necessary for clustering the patents

CPC- Cooperative Patent Classification. The first letter indicates a section that this patent belongs to. In our case, we have 3 of such sections:
A: Human Necessities
G: Physics
H: Electricity
which of course is not surprising. However, when it comes to NLP, we can't just tokenize CPCs in their pure form without getting some type of mess. So, we have to prefix them.
Lets first create three separate columns that would hold the prefixed version of CPCs.

```python
df = df.dropna()
df.reset_index(drop=True, inplace=True)
```

```python
df['cpc_01'] = 'xyz'
df['cpc_02'] = 'xyz'
df['cpc_03'] = 'xyz'
#print(df.cpc[0:10])
for i in range(len(df)):
    if i%100==0:
        print(i)
    df.cpc_01[i] = [first_let[0] for first_let in list(df.cpc.str.split('|').str.join('').str.split()[i])]
    df.cpc_01[i] = re.sub('(\w+)', lambda m: m.group(1) + '_uspto', ' '.join(df.cpc_01[i]))
    df.cpc_02[i] = [rest_let[:] for rest_let in list(df.cpc.str.split('|').str.join('').str.split()[i])]
    df.cpc_03[i] = [five_let[0:5] for five_let in list(df.cpc_02[i])]
    df.cpc_03[i] = re.sub('([\w\d/]+)', lambda m: m.group(1) + '_uspto', ' '.join(df.cpc_03[i]))
    df.cpc_02[i] = re.sub('([\w\d/]+)', lambda m: m.group(1) + '_uspto', ' '.join(df.cpc_02[i]))
```

# Step 5: Clean and lemmatize title, abtract and claims

Before we are able to tokenize and vectorize our data consisted of various strings, we have to make sure that the words we are woring on are in their "purest" form. Otherwise, two words that have the same meaning but are used in different forms (present or past, plural or singular) will not be considered the "same".

```python
def clean_lemmatize(text):
    """
        text: a string
        return: modified initial string with words lemmatized and stopwords/unneeded chars removed
    """
    text = re.sub("[^a-zA-Z]", " ", text)
    text = text.lower().split()
    #remove stopwords after lemmatizing
    text = [lemma.lemmatize(w, 'v') for w in text]
    text = [lemma.lemmatize(w, 'n') for w in text]
    #adding a bunch of symbols to stopwords as they come from cleaning text containing formulas
    stopwords_eng1 = stopwords_eng + ['xn', 'vergleichseinrichtungen','vii','viii','svstem','rck','referenzsignalen','rence', 'renzsignals',     
                          'schiebung','siad','gesamtsignal', 'gung', 'foinied', 'differenzfrequenz','elektronische','entweder',
                            'ferenzsignalen','zugefhrten','sprechenden','strahl', 'ouptut', 'raum', 'weder',   
                            'streufeldes','teilung','trgerfrequenzen','prodicible','foiined','frequenzmoduliert','frequenzsignals',      
                            'frequenzunterschied', 'hervorgerufenen','impprovement', 'interferenzsignalen', 'modulationseinheiten', 'movingfringe',              
                            'multiplikativmen', 'multiplikativer', 'periodischen', 'phasen', 'phasenlagendifferenz','photodetektor',              
                            'photodetektormatrix', 'additionseinrichtung','amplitudenmodulation','amplitudenmodulierten', 'one'
                            'aufteilung','auswertung', 'bereichen','besteht', 'charakteristischen','say','accord', 'bg', 'e', 
                            'fo', 'qcl', 'p', 'thz', 'b', 'q', 'u', 'h', 'f', 'g', 'gf', 'zx', 'zy', 'ff', 'hg', 'z', 'min', 
                            'max', 'w', 'd', 'claim', 'first', 'second', 'wherein', 'two', 'n', 'least', 'measure', 'display', 'input']
    text = [t for t in text if not t in stopwords_eng1]
    return ' '.join(text)
```

```python
df['joined_cleaned'] = 'xyz'
for i in range(len(df)):
    if i%100==0:
        print(i)
    df.joined_cleaned[i] = clean_lemmatize((df.title.astype(str)+ ' ' + df.abstract.astype(str) + ' ' + df.claims.astype(str))[i])
df['joined_all'] = df.joined_cleaned.astype(str) + ' ' + cpcs.values.astype(str) + ' ' + df.cites_01.astype(str) + ' ' + df.examiners.astype(str)
```
